```python
#initial setup
import pandas
import numpy
import matplotlib
%matplotlib inline 
```

```python
#load in data
data = pandas.read_pickle("output.pandas.pkl")
data = data.replace("",numpy.nan)
data
```

```python
list(data.columns)
```

```python
import pickle
keyFile = open("sampleKey.pkl", 'rb')
keyTree = pickle.load(keyFile)
keyFile.close()
keyTree
```

```python
columnNameReplacementTable = {}
for column in data.columns:
    if column.startswith("aMETSIM"):
        newColumnName = column.replace("aMETSIM","")
    elif column.startswith("AT"):
        nameArray = column.split("_")
        run = nameArray[0]
        tag = nameArray[2]
        newColumnName = keyTree[run][tag]
    else:
        raise RuntimeError("You have a column name here that does not fit") #have this crash out
    columnNameReplacementTable[column] = newColumnName    
data.rename(columns = columnNameReplacementTable, inplace=True)
data
```

```python
data.to_csv("metsimRenamed.txt", sep = "\t")
```
