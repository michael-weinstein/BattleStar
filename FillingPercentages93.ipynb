```python
#initial setup
import pandas
import numpy
import matplotlib
%matplotlib inline 
```

```python
#load in data
data = pandas.read_pickle("output.pandas.pkl")
```

```python
#replace all the empty string missing values with NaN
data = data.replace("",numpy.nan)
data
```

```python
#get a general description of the data
description = data.describe()
description
```

```python
#turn the description into a dictionary, keyed to the column names
dictDes = dict(description)
dictDes
```

```python
#get the dimensions of the data
shape = data.shape
shape
```

```python
#get just the number of rows and save that to an appropriate variable
totalRows = shape[0]
totalRows
```

```python
#calculate what percentage of each column (sample) is filled with valid data (and not NaN).  Output will be a list of values and names.
percentValid = []
sampleNames = []
for key in list(dictDes.keys()):
    percentValid.append(dictDes[key][0]/totalRows)
    sampleNames.append(key)
```

```python
#Turn the list of valid fill percentages and names into a pandas dataframe for plotting
validMatrix = pandas.DataFrame(percentValid, sampleNames)
validMatrix.head()
```

```python
#Create a histogram to see what the distribution of filling percentages is
validPlot = validMatrix[0].hist(bins = 100)
validPlot.set_title("BSseq METSIM data by sample")
validPlot.set_xlabel("Percent Filled")
validPlot.set_ylabel("Samples")
```

## Clearly most samples are 90% or more filled in

```python
#Let's do the same thing, looking by locus this time
totalColumns = data.shape[1]
dCount = dict(data.count(axis=1))
locusNames = []
percentages = []
for key in list(dCount.keys()):
    locusNames.append(key)
    percentages.append(dCount[key]/totalColumns)
byLocus = pandas.DataFrame(percentages, locusNames)
byLocus.head()
```

```python
#and we will generate a plot of how much data is filled in for each locus in the samples
validPlot2 = byLocus[0].hist(bins = 50)
validPlot2.set_title("BSseq METSIM data by Locus")
validPlot2.set_xlabel("Percent Filled")
validPlot2.set_ylabel("Loci")
```

##Just under 400K loci are represented in **all** samples
###Will this number improve if I drop samples that have a relatively low percentage of loci filled in?

```python
#deleting any rows with less than 90 percent of loci showing a value
shape = data.shape
totalRows = shape[0]
minFill = int(totalRows * .93)
data.dropna(axis = 1, thresh=minFill, inplace=True)
minFill
```

```python
#check to see if we have removed some columns
data.shape
```

```python
#we did remove columns with the fewest loci filled in.  Let's get ready to reanalyze
description = data.describe()
dictDes = dict(description)
shape = data.shape
shape
totalRows = shape[0]
totalRows
```

```python
#Plot the percent filled by column to be sure that it looks like it is supposed to
percentValid = []
sampleNames = []
for key in list(dictDes.keys()):
    percentValid.append(dictDes[key][0]/totalRows)
    sampleNames.append(key)
validMatrix = pandas.DataFrame(percentValid, sampleNames)
validPlot = validMatrix[0].hist(bins = 20)
validPlot.set_title("BSseq METSIM data by sample")
validPlot.set_xlabel("Percent Filled")
validPlot.set_ylabel("Samples")
```

## We now have only columns where at least 90% of values are present

```python
#Now let's look by row again to see if we have more loci with full datasets
totalColumns = data.shape[1]
dCount = dict(data.count(axis=1))
locusNames = []
percentages = []
for key in list(dCount.keys()):
    locusNames.append(key)
    percentages.append(dCount[key]/totalColumns)
byLocus = pandas.DataFrame(percentages, locusNames)
byLocus.head()
```

```python
#And do the histogram plot
validPlot2 = byLocus[0].hist(bins = 50)
validPlot2.set_title("BSseq METSIM data by Locus")
validPlot2.set_xlabel("Percent Filled")
validPlot2.set_ylabel("Loci")
```
