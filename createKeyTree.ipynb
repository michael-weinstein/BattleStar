```python
#initial setup
import pandas
import numpy
import matplotlib
%matplotlib inline
```

```python
data = pandas.read_csv("keylist.txt", delimiter = "\t")
```

```python
data
```

```python
data.drop(data.index[0])
```

```python
last = ""
for index in range(1,211):
    if pandas.notnull(data["SAMPLE-RUN"][index]):
        last = data["SAMPLE-RUN"][index]
    else:
        data.loc[index, "SAMPLE-RUN"] = last
data
```

```python
keyTree = {}
for index in range(1,211):
    if pandas.isnull(data['SAMPLE'][index]):  #skip any line without a sample ID
        continue
    run = data["SAMPLE-RUN"][index]
    tag = str(int(data["ILLUMINA     TAG"][index]))  #I don't know how there got to be so many spaces in that ID.  Nor do I want to.
    sampleID = str(int(data["SAMPLE"][index]))
    if not run in keyTree:
        keyTree[run] = {}
    keyTree[run][tag] = sampleID
keyTree
```

```python
import pickle
keyFile = open("sampleKey.pkl", 'wb')
pickle.dump(keyTree, keyFile)
keyFile.close()
keyVerify = open('sampleKey.pkl', 'rb')
test = pickle.load(keyVerify)
keyVerify.close()
test
```
